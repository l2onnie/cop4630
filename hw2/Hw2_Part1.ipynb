```
# Import libraries
from mpl_toolkits import mplot3d
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# create the 3 functions
def min(x, y):
  return x**2 + y**2

def max(x, y):
  return -x**2 - y**2

def zero(x, y):
  return x**2 - y**2

x = np.linspace(-6, 6, 60)
y = np.linspace(-6, 6, 60)
X, Y = np.meshgrid(x,y)

# Plot min function
Z = min(X, Y)
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 60, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.scatter(x[30], y[30], alpha=1, color='b', s=300)

# Plot max function
Z = max(X,Y)
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 60, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.scatter(x[30], y[30], alpha=1, color='b', s=300)

# Plot zero vector function
Z = zero(X,Y)
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 60, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.scatter(x[30], y[30], alpha=1, color='r', s=300)
```
